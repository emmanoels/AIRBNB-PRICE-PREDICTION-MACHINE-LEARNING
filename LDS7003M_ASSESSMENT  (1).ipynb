## Filling up missing values

##### 1, Room_private: it is  a boolean column that is, 'TRUE' if the property has a private room and 'FALSE' if the property does not have a private room. We will be using mode because this is the correct method for filling up categorical columns according to (Goyal, C. (2021) ‘How to handle missing values of categorical variables?’, Analytics Vidhya, 04 April,2025. Available at:https://www.analyticsvidhya.com/blog/2021/04/how-to-handle-missing-values-of-categorical-variables/

(Accessed: [16/11/2025]).
 

```python
data['room_private']=data['room_private'].fillna(data['room_private'].mode()[0])
data['host_is_superhost']=data['host_is_superhost'].fillna(data['host_is_superhost'].mode()[0])
```

```python
#from this chart, shows that biz column was originally a categorical column but was converted to float,
# because it has just two distinct values, so for that we use mode to fill up

data['biz'].plot(kind='hist',bins=20, title='buisness ready listing score flag')
```

```python
data['biz']=data['biz'].fillna(data['biz'].mode()[0])
```

```python
data['guest_satisfaction_overall'].plot(kind='hist',bins=20, title='Guest  Satisfaction')
```

```python
data['guest_satisfaction_overall']=data['guest_satisfaction_overall'].fillna(data['guest_satisfaction_overall'].median())
```

```python
data.isnull().sum()
```

```python
data['dist'].plot(kind='hist',bins=20, title='Distance')
```

##### The chart shows the data is normally distributed which means we can use mean to correctly fill the missing values according to Gulati, J., 2025. How to Choose Between Mean, Median, or Mode for Imputation. Machine Learning Mastery. Available at: https://www.statology.org/how-to-choose-between-mean-median-or-mode-for-imputation/
 [Accessed 16 November 2025]. 



```python
data['dist']=data['dist'].fillna(data['dist'].mean())
```

```python
data['metro_dist'].plot(kind='hist',bins=20, title='Distance to nearest metro station')
```

##### The chart above shows the data is left skewed(negatively-skewed) which means most of the data is leaning towards the left, so we make use of median to correctly fill the missing values because it is  robust  according to Gulati, J., 2025. How to Choose Between Mean, Median, or Mode for Imputation. Machine Learning Mastery. Available at: https://www.statology.org/how-to-choose-between-mean-median-or-mode-for-imputation/
 [Accessed 16 November 2025]. 

```python
data['metro_dist']=data['metro_dist'].fillna(data['metro_dist'].median())
```

## Creating new features

```python
data['price_per_person'] = data['realSum'] / data['person_capacity']
data['rooms_per_person'] = data['bedrooms'] / data['person_capacity']
data['total_quality_score'] = data['cleanliness_rating'] + data['guest_satisfaction_overall']
data['is_luxury'] = ((data['bedrooms'] >= 3) &  (data['guest_satisfaction_overall'] > 85)).astype(int)
data['center_proximity_score'] = (1 / (data['dist'] + 1)) + (1 / (data['metro_dist'] + 1))
data['attraction_accessibility'] = data['attr_index_norm'] / (data['dist'] + 1)
data['restaurant_accessibility'] = data['rest_index_norm'] / (data['dist'] + 1)
```

## MODEL DEVELOPMENT AND EVALUATION

## RECURSIVE FEATURE ELIMINATION

```python
X =  data[['room_type', 'person_capacity','guest_satisfaction_overall', 'bedrooms','dist',
       'metro_dist', 'attr_index_norm', 'rest_index_norm', 'lng', 'lat','price_per_person', 'rooms_per_person', 'total_quality_score','center_proximity_score', 'attraction_accessibility']]
y = data['realSum']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2, random_state = 30)
```

```python
from sklearn.ensemble import RandomForestRegressor

model=RandomForestRegressor()
model.fit(X_train,y_train)
```

```python
from sklearn.feature_selection import RFE
rfe = RFE(estimator = model, n_features_to_select = 15)
```

```python
# display selected features
rfe.fit_transform(X, y)
features = X.columns[rfe.support_]
features
```

## Model 1

```python
RFR_prediction=model.predict(X_test)
```

## Evaluation

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
R2score=r2_score(RFR_prediction,y_test)
```

```python
R2score
```

```python
MAE = mean_absolute_error(RFR_prediction, y_test)
MAE
```

```python
MSE = mean_squared_error(RFR_prediction, y_test)
MSE
```

## CROSS VALIDATION

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X, y, cv=5)
print('cross validation score:', scores)
print('cross val mean:', scores.mean())
```

## Grid Search Cv(Hypertunning)

```python
param_grid = {
              'n_estimators': [100, 300],        # number of trees
              'max_depth': [None, 20],        # depth of trees
              'min_samples_split': [2, 5, 10],        # minimum samples to split
              'min_samples_leaf': [1, 2, 4],          # minimum samples per leaf
              'max_features': ['sqrt']
             }

# Open AI(2025)AI-generated code by ChatGPT with the prompt 'give an example of how to perform a grid search cv on random forest regressor'.
# Available at https://chatgpt.com/c/6933f86c-3cdc-832f-92a4-e38aaa93b452. (Accessed: 6th of December,2025).
```

```python
from sklearn.model_selection import GridSearchCV
Grid_Search = GridSearchCV(estimator = model, param_grid = param_grid, cv = 5, scoring = 'accuracy',  n_jobs=-1, verbose=2 )
Grid_Search.fit(X_train, y_train)
```

```python
best_model = Grid_Search.best_estimator_
y_pred = best_model.predict(X_test)
```

```python
# Evaluate the performance of the best model
accuracy = r2_score(y_test, y_pred)
print('R2 score:', accuracy)
```
